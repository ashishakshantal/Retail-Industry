```python
#load the libraries
import pandas as pd
```

```python
#reading the csv file
bbazar = pd.read_excel('bigbazar_dataset.xlsx')
```

```python
bbazar.head()
```

```python
import datetime
year = datetime.datetime.today().year
```

```python
bbazar['year'] = pd.DatetimeIndex(bbazar['transactionDate']).year
bbazar['month'] = pd.DatetimeIndex(bbazar['transactionDate']).month
bbazar['day'] = pd.DatetimeIndex(bbazar['transactionDate']).day
bbazar['quarter'] = pd.DatetimeIndex(bbazar['transactionDate']).quarter
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
bbazar['Name_of_the_day'] = bbazar['transactionDate'].dt.weekday_name
```

```python
bbazar['year'].value_counts().sort_index().plot.bar()
plt.xlabel('Year')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Year')
plt.show()
```

```python
year_2015 = bbazar['year']==2015
year_2016 = bbazar['year']==2016
year_2017 = bbazar['year']==2017
```

```python
bbazar2016 = bbazar[year_2016]
bbazar2017 = bbazar[year_2017]
```

```python
bbazar2015['month'].value_counts().sort_index().plot.bar()
plt.xlabel('Month')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Month for the financial year 2015')
plt.show()

bbazar2016['month'].value_counts().sort_index().plot.bar()
plt.xlabel('Month')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Month for the financial year 2016')
plt.show()

bbazar2017['month'].value_counts().sort_index().plot.bar()
plt.xlabel('Month')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Month for the financial year 2017')
plt.show()
```

```python
bbazar2015['day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by day in a month for the financial year 2015')
plt.show()

bbazar2016['day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by day in a month for the financial year 2016')
plt.show()

bbazar2017['day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by day in a month for the financial year 2017')
plt.show()
```

```python
bbazar2015['Name_of_the_day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days of the Week')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Days in a week for 2015')
plt.show()

bbazar2016['Name_of_the_day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days of the Week')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Days in a week for 2016')
plt.show()

bbazar2017['Name_of_the_day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days of the Week')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Days in a week for 2017')
plt.show()
```
