```python
#load the libraries
import pandas as pd
```

```python
#reading the csv file
bbazar = pd.read_excel('bigbazar_dataset.xlsx')
```

```python
bbazar.head()
```

```python
import datetime
year = datetime.datetime.today().year
```

```python
bbazar['year'] = pd.DatetimeIndex(bbazar['transactionDate']).year
bbazar['month'] = pd.DatetimeIndex(bbazar['transactionDate']).month
bbazar['day'] = pd.DatetimeIndex(bbazar['transactionDate']).day
bbazar['quarter'] = pd.DatetimeIndex(bbazar['transactionDate']).quarter
bbazar['Name_of_the_day'] = pd.DatetimeIndex(bbazar['transactionDate']).day_name()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
bbazar['year'].value_counts().sort_index().plot.bar()
plt.xlabel('Year')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Year')
plt.show()
```

```python
bbazar['month'].value_counts().sort_index().plot.bar()
plt.xlabel('Month')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Month')
plt.show()
```

```python
bbazar['day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by day')
plt.show()
```

```python
bbazar['quarter'].value_counts().sort_index().plot.bar()
plt.xlabel('Quarters')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Quarters')
plt.show()
```

```python
bbazar['Name_of_the_day'].value_counts().sort_index().plot.bar()
plt.xlabel('Days of the Week')
plt.ylabel('Number of transactions')
plt.title('Count of Transactions by Days in a week')
plt.show()
```
